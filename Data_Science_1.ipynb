```python
 import pandas as pd
 import seaborn as sns
 import numpy as np
```

```python
df = pd.read_csv('/content/age_gender.csv',sep=',',  encoding='latin-1')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(df['age'])
plt.title('Total Athletes contribution in summer olympics over time')
plt.xlabel('age')
plt.ylabel('No. of Athlete')
```

```python
athlete_order = df['gender'].value_counts().head(10).index
plt.figure(figsize=(9, 5))
sns.countplot(data=df, y='gender', order=athlete_order)
plt.title('Top 10 Athletes with the most awarded Medals')
plt.xlabel('No. of awrded medals')
plt.ylabel('gender Name');
```

```python
top_10 = df['gender'].value_counts()[:10]
top_10.plot(kind='bar',figsize=(10,8))
plt.title('All Time Medals of top 10 gender')
```

```python
sns.countplot(x='age',hue='gender',data=df)
sns.set(rc={'figure.figsize':(10,10)})
plt.title("Type of genders won over the age")
```

```python
sns.countplot(x='age',data=df)
sns.set(rc={'figure.figsize':(10,10)})
plt.title("Medals won per age")
```

```python
sns.countplot(x="age", hue="gender", data=df)
```

```python
sns.catplot(x="age", y="ethnicity", hue="gender",kind="box", data=df)
```

```python
gender_group = df.groupby(['age', 'gender']).size().unstack()
gender_group.apply(lambda x:x/x.sum(), axis=1).plot(kind='barh', stacked=True, legend=False)
plt.legend(['Men', 'Women'], bbox_to_anchor=(1.0, 0.7))
plt.xlabel('Men / Women ratio')
```

```python
 p = df.groupby('gender').agg('count')
 p.plot(y='age',kind='bar',legend=False,title='Total ages Awarded in each Discipline from 1976-2008',figsize=(20,8))
```

```python
sns.countplot(y='age',hue='age',data=df)
sns.set(rc={'figure.figsize':(10,10)})
plt.xticks(rotation=90)
plt.title('age by gender in each Discipline')
plt.legend(loc=1)  # 1 is code for 'upper right'
```
